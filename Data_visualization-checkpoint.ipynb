```python
import pandas as pd
import seaborn as sns
import numpy as np 
import ipaddress # to get ip address information

from datetime import datetime
from scipy import stats

import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# Getting the file in df, the source csv file is obtained from kaggle.com
cyberattacks= pd.read_csv('Cybersecurity_attacks.csv')
```

```python
# Inspecting the contents of the df
cyberattacks.head()
```

```python
# Checking the size of the df
cyberattacks.shape
```

```python
# Checking the data types of the df
cyberattacks.info()
```

```python
# Dropping an empty column reperesented by a dot
cyberattacks.drop(columns = '.', inplace= True)
```

```python
cyberattacks.columns
```

```python
# Splitting the 'Time' column into 'start_time' and'end_time' then saving them in two different columns
cyberattacks[['start_time','end_time']] = cyberattacks.Time.str.split('-',expand=True)
```

```python
# Checking the structure of the df
cyberattacks.head()
```

```python
# Remove 'Time' column since it is no longer important
cyberattacks.drop(columns='Time',inplace=True)
```

```python
# Checking null values in the df
cyberattacks.isnull().sum()
```

```python
# It is difficult to replace the null values of the attack reference nad those values will create problem in the analysis
# later on if both dropped, so it is preferrable to drop the column since it has no critical information
cyberattacks.drop(columns='Attack Reference', inplace=True) 
```

```python
# Checking the values of 'Attack subcategory' series
cyberattacks['Attack subcategory'].sort_values()
```

```python
# Filtering the number of empty values
cyberattacks.loc[cyberattacks['Attack subcategory']==' ']['Attack subcategory'].count()
```

```python
# Since the number of empty values is not many, it is ok kto drop them
wrong_as= (cyberattacks['Attack subcategory'] == ' ') # save the empty variables to a variable
cyberattacks= cyberattacks[~(wrong_as)].reset_index(drop=True) # drop the empty values from the df
```

```python
# Checking the number of null values in the 'Attack subcategory' series
cyberattacks['Attack subcategory'].isnull().sum()
```

```python
# Since the number is significant, it is good to replace the null values as 'not registered'
cyberattacks['Attack subcategory'] = cyberattacks['Attack subcategory'].fillna('not registered')
```

```python
cyberattacks['Attack subcategory'].isnull().sum()
```

```python
# Changing the 'start_time' and 'end_time' data types from object to int to later change them to datetime 
cyberattacks['start_time']=cyberattacks['start_time'].astype(int)
cyberattacks['end_time']=cyberattacks['end_time'].astype(int)
```

```python
# Changing the int format to datetime format
cyberattacks['start_time'] = [datetime.fromtimestamp(x) for x in cyberattacks.start_time]
cyberattacks['end_time'] = [datetime.fromtimestamp(x) for x in cyberattacks.end_time]
```

```python
# Checking the df contents
cyberattacks.head()
```

```python
# Checking the data types and also check for null values
cyberattacks.info()
```

```python
# Creating additional series 'Attack_duration' to later us it to analyze the df
cyberattacks['Attack_duration']= (cyberattacks['end_time'] - cyberattacks['start_time'])
```

```python
# Checking the newly added series
cyberattacks.head()
```

```python
# Splitting the 'start_time' and 'end_time' columns in to separate date and time columns

cyberattacks['start_date'] = [d.date() for d in cyberattacks['start_time']]
cyberattacks['start_time'] = [d.time() for d in cyberattacks['start_time']]

cyberattacks['end_date'] = [d.date() for d in cyberattacks['end_time']]
cyberattacks['end_time'] = [d.time() for d in cyberattacks['end_time']]
```

```python
# Checking the splitted columns
cyberattacks.head()
```

```python
# Checking the starting date intervals to compare them with ending dates
cyberattacks.start_date.value_counts()
```

```python
cyberattacks.end_date.value_counts()
```

```python
# Since the days are similar, theere is no need to have two columns with the same information,
#so rename the 'start_date' as an 'Attack date'
cyberattacks.rename(columns={'start_date': 'Attack_date'}, inplace=True)
```

```python
# Dropping the redundant series
cyberattacks.drop(columns='end_date', inplace=True)
```

```python
# Checking the new structure of the df
cyberattacks.head()
```

```python
# Checking the total columns and their orders
cyberattacks.columns
```

```python
# Re-ordering the columns to make the df easily understandable
cyberattacks = cyberattacks.reindex(columns= ['Attack category','Attack subcategory','Source IP','Destination IP',
       'Source Port','Destination Port','Protocol','Attack Name','Attack_date',
       'start_time','end_time','Attack_duration'])
```

```python
# Checking the new structure of the df
cyberattacks.head()
```

```python
# Port numbers must be in the range (0,65535), those port number out of this range must be filtered and dropped
wrong_sp = (cyberattacks['Source Port'] < 0) | (cyberattacks['Source Port'] > 65535) 
wrong_dp = (cyberattacks['Destination Port'] < 0) | ( cyberattacks['Destination Port'] > 65535)
```

```python
# Dropping out-of-range port numbers from 'Source Port' and 'Destination Port' columns 
cyberattacks= cyberattacks[~(wrong_sp | wrong_dp)].reset_index(drop=True)
```

```python
# loading the csv file with port description data into a df
tcp_ports= pd.read_csv('TCP_ports.csv')
```

```python
# Checking the contents of the tcp_ports df
tcp_ports.head()
```

```python
# Checking the series with most null values, so the grouping must be done with less number of null values
tcp_ports.isnull().sum()
```

```python
# Merge the two dfs with Port numbers and the additionally required information is the description of the ports
# Save the new df formed by the merge as cyber
cyber = pd.merge(cyberattacks, tcp_ports[['Port','Description']], left_on='Destination Port', right_on='Port', how='left')
```

```python
# Checking hte contents of the new df
cyber.head()
```

```python
# Dropping the redundant series 'Port'
cyber.drop(columns='Port', inplace=True)
```

```python
# Checking the contents of the new df
cyber.head()
```

```python
# Checking how many null values are in the df
cyber.loc[cyber.Description.isnull()]
```

```python
# Checking how mnay null values each port number has
cyber.loc[cyber.Description.isnull()].sort_values(by='Destination Port').groupby("Destination Port").size()
```

```python
# From above, port 514 has the maximum null value. So, save the index numbers of port 514 having UDP protocol
# Giving description for port 514 with UDP protocol from google search
index_514_UDP = cyber.loc[(cyber['Destination Port']==514) & (cyber['Protocol'] == 'UDP')].index
```

```python
# From above, port 514 has the maximum null value. So, save the index numbers of port 514 having TCP protocol
# Giving description for port 514 with TCP protocol from google search
index_514_TCP = cyber.loc[(cyber['Destination Port']==514) & (cyber['Protocol'] == 'TCP')].index
```

```python
# Checking how many null values for port 514 with respective UDP and TCP protocols
cyber.loc[cyber['Destination Port']==514]['Protocol'].value_counts()
```

```python
# From google search, port number 514 has system logging description on UDP
cyber.loc[index_514_UDP, "Description"] = 'used for system logging'
```

```python
# From google search, port number 514 has 'Shell -used to execute non-interactive commands on a remote system' description
cyber.loc[index_514_TCP, "Description"] = 'Shell'
```

```python
# Checking how many null values of 'Description' for different port numbers
cyber.loc[cyber.Description.isnull()].sort_values(by='Destination Port').groupby("Destination Port").size().sort_values(ascending=False)
```

```python
# Checking the content of the df and also checking for missing values
cyber.info()
```

```python
# Since it is difficult to replace all of the null description values with respective description from google search
# and as it creates confusion when analzing, all the null values will be replaced by 'Unassigned' instead of dropping them
cyber.Description.fillna('Unassigned', inplace=True)
```

```python
# Final checking for the contents of the df before visualization
cyber.info()
```

# Visualization

To be included in the visualization

    1) Most frequently used 'Attack category'
    2) Most targeted IP address (destination IP address)
    3) Source IP that launched most of the attack
    4) Most attacked port numbers
    5) Most targeted protocols
    6) Which port has an attack with the longest duration
    7) Which attacks are performed with longer duration
    8) Which port is targeted by specific attack category

```python
#1)  Most frequently used 'Attack category'

plt.figure(figsize=(12,6))
sns.barplot(x=cyber['Attack category'].value_counts().index, y= cyber['Attack category'].value_counts())
plt.xlabel('Attack Category')
plt.ylabel('Count')

plt.title('Number of attacks per attack category')
```

```python
# Showing the most frequent attack categories in pie-chart
#(excluding the two very less frequently used categories)

(cyber['Attack category'].value_counts())[:7].plot(kind='pie',subplots=True, figsize=(8,8))
plt.title('Six most frequent attack categories')
plt.ylabel('')
plt.pie(cyber['Attack category'].value_counts()[:7],autopct='%1.1f%%')
plt.show()
```

```python
# 2) Most targeted IP address (destination IP address)

(cyber['Destination IP'].value_counts()).plot(kind='pie',subplots=True,figsize=(8,8))
plt.title('Most tageted IP addresses')
plt.ylabel('')
plt.pie(cyber['Destination IP'].value_counts(),autopct='%1.1f%%')
plt.show()
```

```python
# 3) Source IP that launched most of the attack

(cyber['Source IP'].value_counts()).plot(kind='pie',subplots=True,figsize=(7,7))
plt.title('Frequently used IP addresses to launch an attack')
plt.ylabel('')
plt.pie(cyber['Source IP'].value_counts(),autopct='%1.1f%%')
plt.show()
```

```python
# 4) Most attacked port numbers

(cyber['Destination Port'].value_counts())[:6].plot(kind='pie',subplots=True,figsize=(8,8))
plt.title('Six most tageted port numbers')
plt.ylabel('')
plt.pie((cyber['Destination Port'].value_counts())[:6],autopct='%1.1f%%')
plt.show()
```

```python
#5) Five most targeted protocols
(cyber['Protocol'].value_counts())[:5].plot(kind='pie',counterclock = True,subplots=True,figsize=(8,8))
plt.title('Five most tageted protocols')
plt.ylabel('')
plt.pie((cyber['Protocol'].value_counts())[:5],autopct='%1.1f%%')
plt.show()
```

```python
#6) Which port has an attack with the longest duration

plt.figure(figsize=(12,6))
sns.stripplot(x='Destination Port',y= 'Attack_duration',data=cyber.loc[cyber['Attack_duration']>30])
plt.title('Port numbers with attack duration in seconds')
plt.show()
```

```python
#7) Which attacks are performed with longer duration

plt.figure(figsize=(12,6))
sns.stripplot(x='Attack_duration',y= 'Attack category',data=cyber.loc[cyber['Attack_duration']>30])
plt.title('Attack categories with their attack duration in seconds')
plt.show()
```

```python
#8) Which port is targeted by specific attack category

plt.figure(figsize=(12,6))
sns.stripplot(x='Attack category',y= 'Destination Port',data=cyber.loc[cyber['Destination Port']<200])
plt.title('Destination ports with common types of attacks')
plt.show()
```
